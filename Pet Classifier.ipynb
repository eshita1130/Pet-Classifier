```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as image
import glob
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
images_fp = './images'
```

```python
image_names = [os.path.basename(file) for file in glob.glob(os.path.join(images_fp,'*.jpg'))]
```

```python
def label_encoder(labels):
    if labels == 'abyssinian': return 0
    elif labels == 'american bulldog': return 1
    elif labels == 'american pit bull terrier': return 2
    elif labels == 'basset hound': return 3
    elif labels == 'beagle': return 4
    elif labels == 'Bengal': return 5
    elif labels == 'Birman': return 6
    elif labels == 'Bombay': return 7
    elif labels == 'boxer': return 8
    elif labels == 'British_Shorthair': return 9
    elif labels == 'chihuahua': return 10
    elif labels == 'Egyptian_Mau': return 11
    elif labels == 'english_cocker_spaniel': return 12
    elif labels == 'english_setter': return 13
    elif labels == 'german_shorthaired': return 14
    elif labels == 'great_pyrenees': return 15
    elif labels == 'havanese': return 16
    elif labels == 'japanese_chin': return 17
    elif labels == 'keeshond': return 18
    elif labels == 'leonberger': return 19
    elif labels == 'Maine_Coon': return 20
    elif labels == 'miniature_pinscher': return 21
    elif labels == 'newfoundland': return 22
    elif labels == 'Persian': return 23
    elif labels == 'pomeranian': return 24
    elif labels == 'pug': return 25
    elif labels == 'Ragdoll': return 26
    elif labels == 'Russian_Blue': return 27
    elif labels == 'saint_bernard': return 28
    elif labels == 'samoyed': return 29
    elif labels == 'scottish_terrier': return 30
    elif labels == 'shiba_inu': return 31
    elif labels == 'Siamese': return 32
    elif labels == 'Sphynx': return 33
    elif labels == 'staffordshire_bull_terrier': return 34
    elif labels == 'wheaten_terrier': return 35
    elif labels == 'yorkshire_terrier': return 36
    
```

```python
features = []
labels = []
IMAGE_SIZE = (224,224)

for names in image_names:
    label = ' '.join(names.split('_')[:-1:])
    label_encoded = label_encoder(label)
    if label_encoded != None:
        img = load_img(os.path.join(images_fp, names))
        img = tf.image.resize_with_pad(img_to_array(img, dtype='uint8'), *IMAGE_SIZE).numpy().astype('uint8')
        image = np.array(img)
        features.append(image)
        labels.append(label_encoded)
```

```python
features_array = np.array(features)
labels_array = np.array(labels)
```

```python
labels_array
```

```python
labels_one_hot = pd.get_dummies(labels_array)
```

```python
labels_one_hot
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train = 65%, val = 25% and test = 20%
x_train, x_test, y_train, y_test = train_test_split(features_array, labels_one_hot, test_size = 0.2, random_state = 42)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.25, random_state = 1)
```

```python
from tensorflow.keras import layers, Input, Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input as pp_i
from tensorflow.keras.layers import RandomFlip, RandomRotation, Dense, Dropout
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
data_augmentation = Sequential([RandomFlip("horizontal_and_vertical"), RandomRotation(0.2)])
prediction_layers = Dense(20, activation = 'softmax')
```

```python
resnet_model = ResNet50(include_top=False, pooling='avg', weights='imagenet')
resnet_model.trainable = False
preprocess_input = pp_i
```

```python
inputs = Input(shape=(224,224,3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = resnet_model(x, training=False)
x = Dropout(0.2)(x)
outputs = prediction_layers(x)
model = Model(inputs, outputs)
```

```python
model.compile(optimizer=Adam(), loss=CategoricalCrossentropy(), metrics = ['accuracy'])
```

```python
model_history = model.fit(x=x_train, y=y_train, validation_data=(x_val, y_val), epochs=10)
```

```python
acc = model_history.history['accuracy']
val_acc = model_history.history['val_accuracy']
loss = model_history.history['loss']
val_loss = model_history.history['val_loss']
```

```python
epochs_range = range(10)

plt.figure(figsize=(15,8))

plt.subplot (1,2,1)
plt.plot(epochs_range, acc, label='Training accuracy')
plt.plot(epochs_range, val_acc, label='Validation accuracy')
plt.legend (loc='upper right')
plt.title('Training and Validation Accuracy')

plt.subplot (1,2,2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss') 
plt. legend (loc='upper right')
plt.title('Training and Validation Loss')
```

```python
model.evaluate(x_test, y_test)
```
